```python
import requests
import matplotlib.pyplot as plt
```

```python
r=requests.get('https://raw.githubusercontent.com/dm-fedorov/python_basic/master/data/opendata.stat')
with open('Data.txt', 'w') as f:
    f.write(r.text)
```

```python
f=open('Data.txt', 'r')
```

```python
pensia = []
date = []

for line in f:
    if line[-1] == '\n':
        line = line[:-1]
    l = line.split(',')
    if l[0] == 'Средняя пенсия' and l[1] == 'Забайкальский край' and '2018' in l[2]:
        date.append(l[2])
        pensia.append(int(l[3]))
```

```python
months = []
for i in range(len(date)):
    month = date[i].split('-')[1]
    months.append(int(month))
```

```python
all_pensia = sum(pensia)
count_pensia = len(pensia)
print(f"Средняя пенсия в 2018 году в Забайкальском крае равна: {all_pensia/count_pensia} рублей.")
```

```python
count_date = len(months)
x = []
y = []
for i in range(count_date):
    x.append(months[i])
    y.append(pensia[i])
```

```python
plt.plot(x, y)
plt.title('Изменение средней пенсии в 2018 году в Забайкальском крае.')
plt.xlabel('Месяц.')
plt.ylabel('Сумма пенсии.')
plt.grid(True)
plt.show()
```
