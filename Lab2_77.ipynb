```python
import requests
```

```python
class MyFile:
    def __init__ (self, name_file, mode):
        self.file_name = name_file
        self.mode = mode
        self.file = None
        self.status = 0
        self.file_url = " "

        if mode == "read":
            self.file = open(name_file, 'r')
        elif mode == "write":
            self.file = open(name_file, 'w')
        elif mode == "append":
            self.file = open(name_file, 'a')
            self.file.write('\n')
        elif mode == "url":
            self.file_name = name_file
        else:
            print(f"Неподдерживаемый режим {mode}.")

    def read(self):
        lines = self.file.readlines()
        result = ''
        for line in lines:
            result+=line
        self.file.close()
        return result
        
    def write(self, text):    
        self.file.write(text)
        self.file.close()

    def read_url(self):
        try:
            response = requests.get(self.file_name, timeout=10)
            response.raise_for_status()
            self.status = 1
            return response.text
        except Exception as e:
            self.status = 0
            return "Ошибка: сайт не существует или нет интернет соединения."
        
    
    def count_urls(self):
        if self.status == 0:
            return "Невозможно получить доступ к сайту."
        else:
            response = requests.get(self.file_name)
            return response.text.count('https://')  
    
    def write_url(self, file_url):
        if self.status == 0:
            return "Невозможно получить доступ к сайту."
        else:
            response = requests.get(self.file_name)
            with open(file_url, 'w') as f:
                f.write(response.text)
    
```

```python
file = MyFile("text.txt", "read")
text = file.read() # происходит чтение в виде str
print(text)
```

```python
file = MyFile("text.txt", "write")
text = file.write("Привет!") # происходит запись строки в файл
```

```python
file = MyFile("text.txt", "append")
text = file.write("привет!") # происходит добавление строки в конец файла
```

```python
# указали URL
file = MyFile("Вставить ссылку", "url")
# и может читать содержимое страницы по указанному URL
text = file.read_url() # происходит чтение в виде str
print(text)
```

```python
# возвращает кол-во url адресов на странице, например, методом count
count = file.count_urls()
print(count)
```

```python
# происходит запись содержимого страницы по URL в указанный файл
file.write_url("text_url.txt")
```
