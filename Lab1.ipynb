```python
import requests
```

```python
r=requests.get('http://dfedorov.spb.ru/python3/sport.txt')
r.encoding = 'cp1251'
with open('sport.txt', 'w', encoding='utf-8') as f:
    f.write(r.text)
```

```python
f=open('sport.txt','r')
k=open('sport.txt', 'r')
```

```python
sports = []


for line in f:
    if line[-1] == '\n':
        line = line[:-1]
    l=line.split('\t')
    if len(l)>1:
        if ',' in l[3]:
            count_sport = l[3].replace(' ', '').split(',')
            for i in range(len(count_sport)):
                if count_sport[i].lower() not in sports:
                    sports.append(count_sport[i].lower())
        else:
            count_sport = l[3].replace(' ', '').lower()
            if count_sport not in sports:
                    sports.append(count_sport)
```

```python
count_object = [0]*len(sports)
```

```python
for line in k:
    if line[-1] == '\n':
        line = line[:-1]
    l=line.split('\t')
    if len(l)>1:
        count_sport = l[3].replace(' ', '').split(',')
        for i in range(len(count_sport)):
            for j in range(1, len(sports)):
                if count_sport[i].lower() == sports[j] and count_sport[i]!='':
                    count_object[j] += 1
```

```python
max1 = max2 = max3 = 0
maxi1 = maxi2 = maxi3 = 0
for i in range(len(count_object)):
    if count_object[i] > max1:
        max3=max2
        max2=max1
        max1=count_object[i]
        maxi3=maxi2
        maxi2=maxi1
        maxi1=i
    elif count_object[i] > max2:
        max3=max2
        max2=count_object[i]
        maxi3=maxi2
        maxi2=i
    elif count_object[i] > max3:
        max3=count_object[i]
        maxi3=i
```

```python
print('Самый популярный в стране спорт: ', (f"{sports[maxi1]},"), 'количество постренных спортивных для него объектов : ', max1)
print('Второй по полярности спорт: ',  (f"{sports[maxi2]},"), 'количество постренных спортивных для него объектов : ', max2)
print('Третий по популярности спорт: ', (f"{sports[maxi3]},"), 'количество постренных спортивных для него объектов : ', max3)
```
